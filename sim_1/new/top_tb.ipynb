```python
from sa import *
SA = SYSTOLIC_ARRAY()
decode = False
```

```python
# 1. Write data to UB
print("1. Write data to UB")
for i in range(256):
    data = encode([i - j for j in range(15, -1, -1)], 16, 8)
    SA.WRITE_DATA(i, data)
data = SA.UB_PRINT(dec=decode)
if (compare("./UB.txt", data) == True):
    print("UB Data is matched.")
```

```python
# 2. Write weight to WB
print("2. Write weight to WB")
for i in range(256):
    weight = encode([- i + j for j in range(15, -1, -1)], 16, 8)
    SA.WRITE_WEIGHT(i, weight)
data = SA.WB_PRINT()
if (compare("./WB.txt", data)):
    print("WB data is matched")
```

```python
# 6. Matrix Multiplication
print("6. Matrix Multiplication")
for i in range(5):
    SA.MAT_MUL(i)
    SA.LOAD_DATA(i)
data = SA.ACC_PRINT(dec=decode)
if (compare("./MMU.txt", data)):
    print("Multiplication result is matched.")
```

```python
# 7. Matrix Multiplication with accumulation
print ("7. Matrix Multiplication with accumulation")
for i in range(5):
    SA.MAT_MUL_ACC(i)
    SA.LOAD_DATA(i)
SA.ACC_PRINT(dec=decode)
print ("DECODE=False")
data = SA.ACC_PRINT(dec=False)
if (compare("./MMU_ACC.txt", data)):
    print("Multiplication with accumulation is matched.")
```
